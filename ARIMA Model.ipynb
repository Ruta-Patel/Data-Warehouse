```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = 18, 8
decomposition = seasonal_decompose(my_df['RESULT'], model='additive')
fig = decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
 
result=adfuller(my_df['RESULT'])
 
#to help you, we added the names of every value
dict(zip(['adf', 'pvalue', 'usedlag', 'nobs', 'critical' 'values', 'icbest'],result))
```

```python

my_df['NEW_RESULT']=my_df['RESULT']-my_df['RESULT'].shift(1)
 
my_df['NEW_RESULT'].plot()
```

```python
my_df['RESULT'].plot()
```

```python
my_df['NEW_RESULT'].head(20)
```

```python
result=adfuller(my_df['NEW_RESULT'].dropna())
dict(zip(['adf', 'pvalue', 'usedlag', 'nobs', 'critical' 'values', 'icbest'],result))
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig1=plot_acf(my_df['NEW_RESULT'].dropna()) 
fig2=plot_pacf(my_df['NEW_RESULT'].dropna()) 
```

```python
my_df['Seasonal_Difference']=my_df['RESULT']-my_df['RESULT'].shift(12)
ax=my_df['Seasonal_Difference'].plot()
```

```python
result=adfuller((my_df['Seasonal_Difference']).dropna())
dict(zip(['adf', 'pvalue', 'usedlag', 'nobs', 'critical' 'values', 'icbest'],result))
```

```python
fig1=plot_acf(my_df['Seasonal_Difference'].dropna()) 
fig2=plot_pacf(my_df['Seasonal_Difference'].dropna()) 
```

```python
import statsmodels.api as sm
pred_model=sm.tsa.statespace.SARIMAX(my_df['RESULT'],order=(1, 1, 1),seasonal_order=(1,1,1,12))
pred_results=pred_model.fit()
my_df['forecast']=pred_results.predict(start=179,end=220,dynamic=True)
my_df[['RESULT','forecast']].plot(figsize=(12,8))
```

```python
pred_results.resid.plot(kind='kde')
```
